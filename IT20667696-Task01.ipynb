```python
# Create random arrays from different distributions
rand_uniform = np.random.rand(100000)
rand_exponential = np.random.exponential(scale=1.0, size=100000)
rand_normal = np.random.normal(size=100000)
```

```python
# Plot the distributions
plt.hist(rand_uniform, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(rand_exponential, density=True, bins=100, histtype="step", color="orange", label="Exponential")
plt.hist(rand_normal, density=True, bins=100, histtype="step", color="green", label="Normal")
```

```python
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
```

```python
# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title("Z = X**2 + Y**2")
plt.show()
```
