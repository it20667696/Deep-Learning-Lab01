```python
import pandas as pd
```

```python
df = pd.read_csv('Au_nanoparticle_dataset.csv')
```

```python
selected_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
```

```python
new_df = df[selected_columns]
```

```python
first20data = df.head(20)
```

```python
print(first20data)
```

```python
mean_values = df.mean()
```

```python
print("\nMean Values:")
print(mean_values)
```

```python
std_dev_values = df.std()
```

```python
print("\nStandard Deviation Values:")
print(std_dev_values)
```

```python
quartile_values = df.describe(percentiles=[0.25, 0.50, 0.75])
```

```python
print(quartile_values)
```
